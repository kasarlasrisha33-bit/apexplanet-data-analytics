```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['Sales'], bins=30)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot(df['Sales'])
plt.title("Sales Boxplot")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.countplot(x='Category', data=df)
plt.title("Category Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Region', data=df)
plt.title("Orders by Region")
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
df['Month'] = df['Order Date'].dt.month

monthly_sales = df.groupby('Month')['Sales'].sum()

plt.figure(figsize=(8,5))
monthly_sales.plot(kind='line')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Month'] = df['Order Date'].dt.month

monthly_sales = df.groupby('Month')['Sales'].sum()

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
monthly_sales.plot(kind='line')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
## Key Insights

1. Sales were lowest in February and highest in November.

2. Sales increased significantly during the last quarter of the year (September–December).

3. November recorded the maximum sales among all months.

4. There is a noticeable drop in sales during October compared to September.

5. Overall sales trend shows growth towards the end of the year, indicating seasonal demand.
```
